```python
from scipy.io.wavfile import read, write
from glob import iglob
from IPython.display import Audio
import matplotlib.pyplot as plt
import numpy as np
import os
import random
%matplotlib inline
```

```python
SAMPLE_RATE = 8000
```

```python
def normalize(data):
    if data.dtype == np.uint8:
        return (data.astype(np.float) - 128) / 128
    elif data.dtype == np.int16:
        return (data.astype(np.float)) / 2 ** 15

def read_wav(path):
    _, data = read(path)
    data = normalize(data)
    if len(data.shape) == 1:
        data = np.reshape(data, data.shape + (1,))
    assert(len(data.shape) == 2)
    return data
```

```python
X = []
y = []
for filename in iglob("free-spoken-digit-dataset/recordings/*.wav"):
    digit, speaker, _ = os.path.basename(filename).split("_")
    data = read_wav(filename)
    if speaker == "nicolas":
        data = data * 3
    for channel in range(data.shape[1]):
        X.append(data[:, channel])
        y.append(int(digit))

n = len(X)
```

```python
def random_seq(length, silence_from=0.5, silence_to=1.0, noise_db=-10):
    seq_parts = []
    y_seq_parts = []
    seq_parts.append(np.zeros(int(SAMPLE_RATE * np.random.uniform(silence_from, silence_to))))
    for i in range(length):
        selected_id = np.random.choice(n)
        seq_parts.append(X[selected_id])
        seq_parts.append(np.zeros(int(SAMPLE_RATE * np.random.uniform(silence_from, silence_to))))
        y_seq_parts.append(y[selected_id])
    audio = np.concatenate(seq_parts)
    signal_rmse = np.sqrt(np.var(audio))
    audio += np.random.normal(scale = 10 ** (noise_db / 20) * signal_rmse, size=audio.shape)
    return audio, y_seq_parts
    
```

```python
audio, true_labels = random_seq(10, noise_db=-50, silence_to=1.0, silence_from=0.3)
print(true_labels)
Audio(data=audio, rate=SAMPLE_RATE)
```
